```python
import pandas as pd
import geopandas as gpd
from pandas import DataFrame
```

```python
#メッシュデータ
gdf_mesh = gpd.read_file('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/奥州市　重心/gdf_osyu_100m_mesh.shp')
gdf_mesh.rename(columns={'longitude':'lon', 'latitude':'lat', 'geometry':'geom'}, inplace=True)
gdf_mesh = gdf_mesh[['Meshcode', 'geom', 'lat', 'lon', 'lu_code', 'lu_year']]
```

```python
#標高
gdf_elevation = pd.read_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/奥州市　標高/奥州市_標高.csv')
gdf_elevation.rename(columns={'hyoukouchi_mean':'elevation_avg', 'hyoukouchi_max':'elevation_max', 'hyoukouchi_min':'elevation_min', 'hyoukouchi_max-hyoukouchi_min':'elevation_delta'}, inplace=True)
gdf_elevation.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#傾斜
gdf_slope = pd.read_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/src/数値地図/奥州市　傾斜/03215_6678_SLOPE.csv', encoding='shift-jis')
gdf_slope['slope_delta'] = gdf_slope['slopemax'] - gdf_slope['slopemin']
gdf_slope.rename(columns={'slopemean':'slope_avg', 'slopemin':'slope_min', 'slopemax':'slope_max'}, inplace=True)
gdf_slope = gdf_slope[['Meshcode', 'slope_avg', 'slope_max', 'slope_min', 'slope_delta']]
```

```python
#天気
gdf_weather = pd.read_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/奥州市　天気/奥州市_気候.csv')
gdf_weather.rename(columns={'G02_053':'temp_avg', 'max_temperature':'temp_max', 'min_temperature':'temp_min', 'avg_rainfall':'rainfall_avg', 'max_rainfall':'rainfall_max'}, inplace=True)
gdf_weather.drop(['Unnamed: 0', 'geometry'], axis=1, inplace=True)
```

```python
#道路点検
gdf_road = pd.read_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/奥州市　道路舗装/奥州市_道路舗装.csv', encoding='shift-jis')
gdf_road.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#騒音
gdf_noise = pd.read_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/奥州市　騒音/奥州市_騒音.csv')
gdf_noise.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#建物密集
gdf_building = pd.read_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/奥州市　建物密集度/奥州市_建物密集度.csv')
gdf_building.rename(columns={'building_ratio':'bldg_density'}, inplace=True)
gdf_building.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#道路幅員
gdf_width = pd.read_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/奥州市　道路幅員/road_width.csv', encoding='utf-8')
gdf_width.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#最近傍道路
gdf_nearroad = pd.read_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/奥州市　最近傍主要道路/奥州市_主要道路.csv', encoding='shift-jis')
```

```python
#最近傍高速
gdf_motoway = pd.read_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/奥州市　最近傍購高速道路/奥州市_最近傍高速道路.csv', encoding='shift-jis')
```

```python
#最近傍線路
gdf_railway = pd.read_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/奥州市　最近傍鉄道/奥州市_最近傍線路.csv', encoding='shift-jis')
```

```python
#最近傍駅
gdf_station = pd.read_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/奥州市　最近傍鉄道/奥州市_最近傍駅.csv', encoding='shift-jis')
```

#結合

```python
#データフレームを結合するためリストに格納
gdf_frames = [gdf_elevation, gdf_slope, gdf_weather, gdf_road, gdf_noise, gdf_building, gdf_width, gdf_nearroad, gdf_motoway, gdf_railway, gdf_station]

n = 1
for i in gdf_frames:
  if n == 1:
    gdf_merge = pd.merge(gdf_mesh, i,  on='Meshcode')

  if n >= 2:
    gdf_merge = pd.merge(gdf_merge, i,  on='Meshcode')

  n += 1

gdf_merge.rename(columns={'Meshcode':'meshcode'}, inplace=True)
```

```python
gdf_merge.to_csv('/content/drive/MyDrive/水道管劣化予測/水道管劣化予測/result/奥州市/データ結合/奥州市_周辺環境データ.csv', encoding='shift-jis', index=False)
```
