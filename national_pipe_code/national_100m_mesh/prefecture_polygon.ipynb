```python
import geopandas as gpd
import pandas as pd
from collections import OrderedDict
from shapely.ops import unary_union
```

```python
prefecture_path = 'I:/マイドライブ/水道管劣化予測/水道管劣化予測/src/国土数値情報/土地利用メッシュ/mesh_for_work/prefecture_polygon/N03-20230101_GML/N03-23_230101.shp'
prefecture_polygon = gpd.read_file(prefecture_path, encoding='shift-jis')
```

```python
#座標系の変換
dst_proj = 4326  # 変換後の座標系を指定
prefecture_polygon = prefecture_polygon.to_crs(epsg=dst_proj)  # 変換後座標に変換
```

```python
#元のdfの順序は保持する
pref_list = list(OrderedDict.fromkeys(prefecture_polygon['N03_001']))
```

```python
#県のポリゴン作成
pref_gdf = gpd.GeoDataFrame(columns=['pref', 'geometry'])

for pref in pref_list:
    #県ごとのポリゴン作成
    pref_polygon = prefecture_polygon[prefecture_polygon['N03_001'] == pref]['geometry'].unary_union
    
    pref_tenmp_gdf = gpd.GeoDataFrame({
            'pref':[pref],
            'geometry':[pref_polygon],
        })
    
    pref_gdf = pd.concat([pref_gdf, pref_tenmp_gdf], ignore_index=True)
```

```python
#座標変換
if pref_gdf.crs is None:
	pref_gdf.set_crs(epsg=4326, inplace=True)
```

```python
pref_polygon_path = 'I:/マイドライブ/水道管劣化予測/水道管劣化予測/src/国土数値情報/土地利用メッシュ/mesh_for_work/prefecture_polygon/pref_polygon.shp'
pref_gdf.to_file(pref_polygon_path, encoding='shift-jis')
```
